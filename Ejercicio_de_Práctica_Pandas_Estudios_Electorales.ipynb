### 2. Procesamiento de datos

Queremos ajustar el conjunto de datos para nuestro uso. Haz lo siguiente:

- Cambie el nombre de la columna `educ` a `education`.
- Cree una nueva columna llamada `party`, basada en la respuesta de cada encuestado al `PID`. `party` debe ser igual a `Democrat` si el encuestado seleccionó Demócrata fuerte o Demócrata débil. `party` será igual a `Republican` si el encuestado seleccionó Republicano Fuerte o Débil para `PID` e `Independent` si seleccionó cualquier otra cosa.
- Cree una nueva columna llamada `age_group` que clasifique a los encuestados en las siguientes categorías según su `age`: 18-24, 25-34, 35-44, 45-54, 55-64 y 65 y más.

```python
# Código aquí
df.rename(columns={'educ': 'education'}, inplace=True)
df
```

```python
def party(PID):
    if PID in [1,0]: return "Democrat"
    elif PID in [5,6]: return "Republican"
    else: return "Independiente"

df["Party"] = df["PID"].apply(party)
df
```

```python
def categorias_edad(age):
    if age >= 18 and age <= 24:
        return '18-24'
    elif age >= 25 and age <= 34:
        return '25-34'
    elif age >= 35 and age <= 44:
        return '35-44'
    elif age >= 45 and age <= 54:
        return '45-54'
    elif age >= 55 and age <= 64:
        return '55-64'
    else:
        return '65 y más'

df['age_group'] = df['age'].apply(categorias_edad)
df
```

### 4. Cálculos a partir de los datos

Para cada uno de los siguientes enfrentamientos, elija el grupo que tenga más probabilidades de votar por Bill Clinton. Puedes calcular esto usando el porcentaje de cada grupo que tiene la intención de votar por Clinton (`votar`). ¿Qué enfrentamiento fue el más cercano? ¿Cuál tuvo la mayor diferencia?

Otra forma de pensar en esto: dado que un encuestado es demócrata, hay un ____ por ciento de posibilidades de que vote por Clinton. ¿Cómo cambia este valor si el encuestado es republicano?

* Personas que ven noticias de televisión al menos 6 días a la semana o Personas que ven noticias de televisión menos de 3 días a la semana
* Personas que viven en algún lugar con una población mayor que el encuestado promedio o Personas que viven en un lugar con una población igual o menor que el encuestado promedio

```python
# Código aquí
def voting_percentage_Clinton(group):
  return (group['vote'] == 0).mean() * 100
```

```python
df_democrats = df[df.party == 'Democrat']
total_democrats =  len(df_democrats)
df_democrats_Clinton = df_democrats[df_democrats.vote == 0]
Clinton_democrats =  len(df_democrats_Clinton)
p_democrats = Clinton_democrats / total_democrats *100

df_republicans = df[df.party == 'Republican']
total_republicans =  len(df_republicans)
df_republicans_Clinton = df_republicans[df_republicans.vote == 0]
Clinton_republicans =  len(df_republicans_Clinton)
p_republicans = Clinton_republicans / total_republicans *100

print('% de Democratas que votarían por Clinton:', round(p_democrats,2),'%')
print('% de Republicanos que votarían por Clinton:',round(p_republicans,2),'%')
```

```python
prob_clinton_menor44 = len(df[(df['age']<=44)&(df['vote']==0)])/len(df['age']<=44)
prob_clinton_mayor44 = len(df[(df['age']>44)&(df['vote']==0)])/len(df['age']>44)
print(f'La probabilidad de que alguien menor a 44 años vote por clinton es de {round(prob_clinton_menor44*100,2)}%')
print(f'La probabilidad de que alguien mayor a 44 años vote por clinton es de {round(prob_clinton_mayor44*100,2)}%')
```
